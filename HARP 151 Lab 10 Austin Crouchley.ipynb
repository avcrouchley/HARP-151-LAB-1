```python
from tkinter import *
import requests
import tkinter as tk

genius = requests.get(f"https://api.quotable.io/authors")
json_file = genius.json()
results = json_file["results"]

x = 0
names = []
for i in results: 
    one = results[x]
    name = one["name"]
    names.append(name)
    x += 1

root = Tk()

def getquote(): 
    author = value.get()
    genius = requests.get(f"https://api.quotable.io/quotes/random?author={author}")
    json_file = genius.json()
    quote = json_file[0]["content"]
    myLabel1 = Label(root, text=quote)
    myLabel1.pack()

value = StringVar()
value.set("Select an Author")
    
dropdown = OptionMenu(root, value, *names)
choose_button = Button(root, text="Click to choose", command=getquote)

dropdown.pack()
choose_button.pack()

root.mainloop()
```

```python
import requests
import json
import tkinter as tk
from tkinter import *

root = Tk()

lat = "42.0987"
lon = "-75.9180"
genius = requests.get(f"https://api.weather.gov/points/{lat},{lon}")
json_file = genius.json()
       
forecast = json_file["properties"]["forecast"]
new_request = requests.get(forecast)

json = new_request.json()

a = json["properties"]["periods"]
a

for i in a: 
    name = i["name"]
    temp = i["temperature"]
    detail = i["detailedForecast"]
    myLabel1 = Label(root, text=name)
    myLabel1.pack()
    myLabel2 = Label(root, text=temp)
    myLabel2.pack()
    myLabel3 = Label(root, text=detail)
    myLabel3.pack()
    
    
root.mainloop()
```
